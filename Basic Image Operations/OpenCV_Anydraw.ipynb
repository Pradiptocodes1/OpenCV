```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def showit(Title="",image=None,size=10):
  h,w = image.shape[0],image.shape[1]
  aspect_ratio=h/w
  plt.figure(figsize=(size*aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(Title)
  plt.show()
```

```python
black_image=np.zeros((512,512,3),np.uint8)
grey_image=np.zeros((512,512),np.uint8)
showit("Black Image",black_image,size=4)
showit("Grey Image",grey_image,size=4)
```

```python
cv2.line(black_image,(200,250),(400,450),(255,127,0),3) #directly alters the input image
cv2.line(black_image,(200,450),(400,250),(234,90,69),3)
cv2.rectangle(black_image,(200,250),(400,450),(229,69,1),3)
cv2.rectangle(black_image,(200,100),(400,100),(0,56,190),3)
cv2.circle(black_image,(350,350),50,(0,56,190),-1)
pts=np.array([[200,250],[250,100],[200,400],[400,300]],np.int32)
pts=pts.reshape((-1,1,2))
cv2.polylines(black_image,[pts],True,(0,189,0),3)
sometext='Hi There!'
cv2.putText(black_image,sometext,(200,80),cv2.FONT_HERSHEY_SCRIPT_COMPLEX,3,(0,0,90),5)
showit("AnyDraw",black_image,size=3)
```
