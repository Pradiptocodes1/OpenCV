```python
images=cv2.imread('./images/flowers.jpeg')
```

```python
import matplotlib.pyplot as plt
plt.imshow(cv2.cvtColor(images,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
def showit(title="", images=None):
  plt.imshow(cv2.cvtColor(images,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
showit("image1",images)
```

```python
cv2.imwrite("output.png",images)
```

```python
def showdim(image=None):
  print('Displaying Image Dimensions--')
  print('Height: {}px'.format(int(image.shape[0])))
  print('Width: {}px'.format(int(image.shape[1])))
  print('Depth: {}px'.format(int(image.shape[2])))
```
