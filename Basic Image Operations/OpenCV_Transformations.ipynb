```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def showit(Title="",image=None,size=10):
  w,h=image.shape[0],image.shape[1]
  aspect_ratio=w/h
  plt.figure(figsize=(size*aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(Title)
  plt.show()
```

```python
image = cv2.imread('images/Trump.jpg')
showit("Image", image,size=5)
```

```python
h,w = image.shape[:2]
qtr_h,qtr_w = h/4,w/4
T=np.float32([[1,0,qtr_h],[0,1,qtr_h]])
image_translated=cv2.warpAffine(image,T,(w,h))
showit("Translated Image",image_translated,size=5)
```

```python
image=cv2.imread("images/Trump.jpg")
rotation_matrix=cv2.getRotationMatrix2D((w/2,h/2),90,1)#here 1 is the scale
Image_rotated=cv2.warpAffine(image,rotation_matrix,(w,h))
showit("Image_rotated",Image_rotated,size=5)
```

```python
image=cv2.imread("images/Trump.jpg")
rotated_using_transpose=cv2.transpose(image)#transpose is less flexible
showit("Transpose rotation",rotated_using_transpose,size=5)
```

```python
flipit=cv2.flip(rotated_using_transpose,1)
showit("Flipped Image",flipit,size=5)
```
