```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def showit(Title="",image=None,size=10):
  w,h=image.shape[0],image.shape[1]
  aspect_ratio=w/h
  plt.figure(figsize=(aspect_ratio*size,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(Title)
  plt.show()
```

```python
image=cv2.imread('images/liberty.jpeg',0)#comma zero indicated grayscale
showit("Sample Image",image)
print(image)
M=np.ones(image.shape,dtype="uint8")*100
print(M)
```

```python
bright_image=cv2.add(image,M)
showit("Bright Image",bright_image)
```

```python
added2 = image + M 
showit("Simple Numpy Adding Results in Clipping", added2)
#in this case we will be getting irregular image since in some case the value might increase way beyond the limit of 255.
```

```python
decreased_brightness=cv2.subtract(image,M)
showit("Decreased Brightness",decreased_brightness)
```

```python
square = np.zeros((300, 300), np.uint8)
cv2.rectangle(square, (50, 50), (250, 250), 255, -2)
showit("square", square)
ellipse = np.zeros((300, 300), np.uint8)
cv2.ellipse(ellipse, (150, 150), (150, 150), 30, 0, 180, 255, -1)
showit("ellipse", ellipse)
```

```python
And=cv2.bitwise_and(square,ellipse)
showit("And operator",And)
OR=cv2.bitwise_or(square,ellipse)
showit("OR operator",OR)
XOR=cv2.bitwise_xor(square,ellipse)
showit("XOR operator",XOR)
NOT=cv2.bitwise_not(ellipse)
showit("NOT operator",NOT)
```
